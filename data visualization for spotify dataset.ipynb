```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
spotify_data = pd.read_csv("spotify-2023.csv", encoding='ISO-8859-1')
print("1. What are the top 5 songs and artists released in 2023?")
topFiveSongs = spotify_data['track_name'].head(5)      # how to specifically access released in 2023
topFiveArtists = spotify_data['artist(s)_name'].head(5)
songs = pd.DataFrame(spotify_data) 

# print(topFiveSongs)
# print(topFiveArtists)

print("shows the number of artists in the top 5 songs")
seaborn.barplot(x=spotify_data['artist_count'], y=topFiveSongs,data= songs )

plt.show() # shows number of artists in each song

# 2. What are the 5 least popular songs and artists released in 2023?
LeastFiveSongs = spotify_data['track_name'].tail(5)    #same thing
LeastFiveArtists = spotify_data['artist(s)_name'].tail(5)
songs = pd.DataFrame(spotify_data) 
# print(LeastFiveSongs)
# print(LeastFiveArtists)
print("shows the number of artists in the 5 least popular songs")
seaborn.barplot(x=spotify_data['artist_count'], y=LeastFiveSongs,data= songs )
plt.show()

# 3. What are the top 10 most popular songs of 2022?
mostPopular = spotify_data[['track_name'],spotify_data['released_year'] == '2022'].head(10)
print(mostPopular)
```
