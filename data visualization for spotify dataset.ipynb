```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
spotify_data = pd.read_csv("spotify-2023.csv", encoding='ISO-8859-1')

print("1. What are the top 5 songs released in 2023?")
topFiveSongs = spotify_data[spotify_data['released_year'] == 2023]['track_name'].head(5)   # how to specifically access released in 2023
streams = spotify_data[spotify_data['released_year'] == 2023]['streams'].head(5)
# topFiveArtists = spotify_data['artist(s)_name'].head(5)
songs = pd.DataFrame(spotify_data) 

# print(topFiveSongs     debugging
# print(topFiveArtists)

print("   shows the number of artists in the top 5 songs")
seaborn.barplot(x = streams, y = topFiveSongs )

plt.show() # shows number of artists in each song

```

The top 5 songs released in 2023 are Seven, LALA, Where She Goes, and Sprinter.

```python
print('2. What are the 5 least popular songs released in 2023?')
LeastFiveSongs = spotify_data[spotify_data['released_year'] == 2023]['track_name'].tail(5)    #same thing
leastStreams = spotify_data[spotify_data['released_year'] == 2023]['streams'].tail(5) 
# print(LeastFiveSongs)     debugging
# print(LeastFiveArtists)
print("shows the number of artists in the 5 least popular songs")
seaborn.barplot(x=leastStreams, y=LeastFiveSongs )
plt.show()
```

The 5 least popular songs released in 2023 are I'm Not Here To Make Friends, TRUSTFALL, VIBE, Mui, and Sugar Rush Ride. 

```python
print('3. What are the top 10 most popular songs of 2022?')
mostPopular = spotify_data[spotify_data['released_year'] == 2022]['track_name'].head(10)
streams = spotify_data[spotify_data['released_year'] == 2022]['streams'].head(10)
# print(mostPopular)  debugging
plt.title('Tops songs of 2022')
plt.pie(streams, labels = mostPopular)
plt.show()
```

The top 10 most popular songs of 2022 are As It Was, Hey Mor, Left and Right, La Bachata, I Ain't Worried, I'm Good, Anti-Hero, Creepin', Calm Down, and Kill Bill.
